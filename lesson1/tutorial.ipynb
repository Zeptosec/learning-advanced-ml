## Creating data
We are going to create an array of values for the X axis so that we can calculate the Y axis for the __sin__ and __tanh__ functions

```python

import numpy as np

CYCLES = 4
SIZE = 100

x = np.linspace(0, 2 * np.pi * CYCLES, SIZE)
tanhy = np.tanh(x)
siny = np.sin(x)
```

## Displaying
Now that we have all the data we can display it using matplotlib library.

```python

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'hspace': 0.6, 'wspace': 0.9})
ax1.plot(x, tanhy)

ax1.set(xlabel='X axis', ylabel='Y axis',
       title='f(x) = tanh(x)')
ax1.grid()

ax2.plot(x, siny)

ax2.set(xlabel='X axis', ylabel='Y axis',
       title='f(x) = sin(x)')
ax2.grid()

plt.show()
```
